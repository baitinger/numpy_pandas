```python
boston = load_boston()
```

```python
data = boston["data"]
```

```python
feature_names = boston["feature_names"]
```

```python
target = boston["target"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
```

```python
y = pd.DataFrame(target, columns=['price'])
```

Далее стандартизирую признаки, перед разделением на тренировочные и валидационные данные. 


```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

Далее хотелось посмотреть на получившиеся предсказания, чтобы хоть как-то понять получились предсказания или нет.

```python
check_test = pd.DataFrame({"y_test":y_test['price'], "y_pred":y_pred.flatten()})

check_test.head(15)
```

```python
from sklearn.metrics import r2_score
```

```python
test_r2 = r2_score(y_test['price'], y_pred.flatten())

test_r2
```
